# Прогнозирование эффективности обогащения концентрата

**Описание задачи:**  
Для разработки модели предсказания коэффициента восстановления золота из золотосодержащей руды предоставлены данные с параметрами добычи и очистки. Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками. У признаков названия строятся по правилу: `этап`.`тип_параметра`.`название_параметра`

**Цель исследования:**  
По предоставленным данным о процессе восстановления золота из золотосодержащей руды нужно разработать модель машинного обучения, которая лучше всего спрогнозирует две величины:

В качестве метрики качества выступает `sMAPE` (25% метрики для чернового, 75% для финального), обучающий и тестовый датасет уже предоставлены.

#### Проверка корректности значений целевых признаков

Нужно проверить, правильно ли расчитаны целевые признаки в общем датасете. Для этого вычислю `MAE` для значений датасета и значений, которые получу по формуле.

```python
def recovery_value(data, final = False):
    if final: # этап очистки
        step = 'final'
    else: # этап флотации
        step = 'rougher'
    
    feed = data['rougher.input.feed_au']
    concentrate = data[step + '.output.concentrate_au']
    tail = data[step + '.output.tail_au']
    
    
    return (concentrate * (feed - tail)) / (feed * (concentrate - tail)) * 100
```

Значения `MAE` (сумма модулей разности настоящих и предсказанных значений) для целевых признаков практически нулевые, значит значения в датасете соответствуют действительности.

```python
mean_absolute_error(df[TRAIN]['final.output.recovery'], recovery_value(df[TRAIN], True))
```

```python
mean_absolute_error(df[TRAIN]['rougher.output.recovery'], recovery_value(df[TRAIN], False))
```

Функции возвращают таблицу с описательными характеристиками указанных признаков и отрисовывают график для каждого признака.

### Объявление функций

```python
def show_distr(data, group, draw_plot=True, set_main_title='?', bins_set=np.linspace(0, 50, num=25), titles='', plot_attr=None):
    if type(data) != list:
        data = [data]
    if type(group) != list:
        group = [group]
    if plot_attr == None:
        plot_attr = {}
    
    if draw_plot:
        col = 2
        row = 1 + (len(group) * len(data) // col)
        
        if len(group) == 1:
            col = 1
        elif (len(group) % col) == 0:
            row -= 1
        fig, axes = plt.subplots(row, col, figsize=(col*8, row*6), sharey=True, squeeze=False)
        fig.suptitle(set_main_title, fontsize = 30)
    
    table = pd.DataFrame()
    
    for ind, dataset in enumerate(data):
        for i, columns in enumerate(group):
            if type(columns) == str:
                column_comp = df[FULL].loc[:,df[FULL].columns.str.contains(columns, regex = True)].columns
            else:
                column_comp = columns
            
            desc = dataset.loc[:,column_comp].describe()
            if (len(table) == 0):
                table = desc
            else:
                table = table.merge(desc, left_index = True, right_index = True, suffixes = ('_0', '_1'))
    
            if draw_plot:
                ax_set = axes[i // col][i % col]
                sns.histplot(dataset.loc[:,column_comp], bins = bins_set, ax=ax_set, **plot_attr)
                ax_set.set_ylabel('Количество', fontsize = 15)
                ax_set.set_xlabel('Значение', fontsize = 15)
                try:
                    ax_set.set_title(titles[i], fontsize = 22)
                except:
                    pass
    
    if draw_plot:
        plt.tight_layout()
    
    return table
```

## Обучение и тестирование моделей

### Объявление функций

`sMAPE` (симметричное среднее абсолютное процентное отклонение) выступает метрикой качества для выявления лучшей модели. Чем меньше значение метрики, тем выше качество модели.

```python
def smape(y, y_pred):
    y = np.array(y)
    n_targ = y.shape[1]
    length = y.shape[0]
    score = list(range(0, n_targ))
    for i in range(0, n_targ):
        cur_targ = y[:,i]
        cur_pred = y_pred[:,i]
        score[i] = np.sum(np.abs(cur_targ - cur_pred) / (np.abs(cur_targ) + np.abs(cur_pred))) * 200/length
        
    return 0.25*score[0] + 0.75*score[1]

smape_scorer = make_scorer(smape, greater_is_better=False)
```

Функция `create_pipeline` создает пайплайн, выводит результаты обучения.

```python
def create_pipeline(regressor, trans_set=None, search_type='grid', name='unknown',
                    cv=5, set_iter=10, grid = None, fit_params=None):
    result = {}
    trans = []
    
    if trans_set != None:
        trans.append(('trans', trans_set, cols))
    grid = {} if grid == None else grid
    fit_params = {} if fit_params == None else fit_params
    
    if search_type == 'grid':
        model = GridSearchCV(regressor, param_grid=grid, scoring=smape_scorer, cv=cv, n_jobs=-1)
    elif search_type == 'random':
        model = RandomizedSearchCV(regressor, param_grid=grid, scoring=smape_scorer, n_iter=set_iter, cv=cv, n_jobs=-1)
        
    step = []
    if (len(trans) > 0):
        step.append(('preprocesser', ColumnTransformer(transformers=trans, remainder='passthrough')))
    step.append(('regressor', model))
    
    pipe = Pipeline(steps=step)
    pipe.fit(features_train, target_train, **fit_params)
    pipe.name_ = name
    pipe.named_steps.regressor.best_score_ = -1 * pipe.named_steps.regressor.best_score_
    
    print(f'Score {pipe.name_}: {pipe.named_steps.regressor.best_score_}')
    print(f'Params {pipe.name_}: {pipe.named_steps.regressor.best_params_}')
    
    return pipe
```

Функция `data_split` формирует тренировочную и тестовую выборку, убирает из выборок указанные столбцы.

```python
global features_train, features_test, target_test, target_train, cols
```

```python
def data_split(drop):
    global features_train, features_test, target_test, target_train, column_drop, cols
    
    total_drop = list(set(drop + column_drop))
    
    target_train = df_cut[TRAIN][['rougher.output.recovery', 'final.output.recovery']].reset_index(drop = True).copy()
    features_train = df_cut[TRAIN].drop(columns=total_drop).reset_index(drop = True).copy()
    target_test = (
        df_cut[FULL].loc[df_cut[FULL].index.isin(df_cut[TEST].index),['rougher.output.recovery', 'final.output.recovery']]
        .reset_index(drop = True).copy()
    )
    features_test = df_cut[TEST].drop(columns=drop).reset_index(drop=True).copy()
    cols = features_train.columns
    
    print('Train shape:', features_train.shape)
    print('Test shape:', features_test.shape)
    skim(features_train)
```
