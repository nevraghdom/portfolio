# Прогноз оттока клиентов у оператора связи

## Введение

**Описание задачи:** Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов; если пользователь планирует уйти, компания предложит ему промокоды и специальные условия. Были собраны персональные данные о некоторых клиентах, информацию об их тарифах и договорах, вся информация актуальна на 1 февраля 2020 года. Клиенты могут воспользоваться двумя основными услугами: интернет и телефонная связь.  

**Цели исследования:**  
Необходимо построить модель машинного обучения, которая будет прогнозировать уйдет клиент в ближайшее время или нет. В распоряжении персональные данные о некоторых клиентах, информация об их тарифах и договорах:  
- Целевой признак - уход клиента, задача бинарной классификации.
- Метрикой выступает `ROC-AUC`. Значение метрики должно быть не ниже 0.78, желательно значение не ниже 0.85.
- За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.
- Данные хранятся в нескольких таблиах, ключом для их объединения является признак `customerID`.

**Ход исследования:**
1. **Обзор данных.**
    - Чтение и проверка данных, объединение датафреймов.
    - Приведение признаков к нужному виду, изменение имен признаков, определение столбцов с пропущенными значениями.
    - Выделение целевого признака и метрик качества.
2. **Исследовательский анализ данных**
    - Проверка признаков на линейную корреляцию, заполнение пропущенных значений.
    - Изучение типа распределений признаков, обнаружение потенциальных взаимосвязей, удаление ненужных наблюдений/признаков, если таковые имеются.
3. **Обучение моделей**
    - Разделение выборки на тренеровочную и тестовую, выбор типов моделей, подходящих для решения поставленной задачи. Подготовка пайплайна.
    - Обучение моделей с использованием кросс-валидации, фиксирование значений метрик.
4. **Тестирование моделей**
    - Проверить обученные модели на адекватность.
    - Тестирование моделей на выделенной тестовой выборке, которая не участвовала в обучении, сравнение моделей по метрикам.
5. **Интерпретация результата и общий вывод**
    - Выбор лучшей модели/моделей с обоснованием выбора, фиксация основных моментов проекта в общем выводе, помимо промежуточных.

### Объявление функций

Функция `draw_hist` выводит график распределения значений относительно целевого признака. Статистика для категориальных признаков:  
- `count` - количество вхождений
- `percent` - доля `count` в общем количестве наблюдений
- `percent_group` - доля `count` для выделенной группы признака `column` (проверка по признаку)
- `percent_customer` - доля `count` в общем количестве наблюдений для этой группировки (проверка по `customer_left`)

```python
def draw_hist(column, query=None,
              set_ticks='percent_customer', ticks_rotation=0,
              ylabel='Count', xlabel=None, set_title='Feature distribution',
              plot_attr=None):
    
    data_query = data
    
    if query != None:
        data_query = data.query(query)
        column = column[data_query.index]
        
    target_col = 'customer_left'
    target = data_query[target_col].astype('int')
    
    category = column.dtype in ['string', 'category', 'O', 'object']
    plot_attr = {} if plot_attr == None else plot_attr
    plot_attr.setdefault('multiple', 'stack')
    dfs = [column[target == 0], column[target == 1], column]
    df_names = ['Here', 'Left', 'Total']
    df_concat = []
    agg_func = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std']
    ticks_change = False
    alpha = 0.05
    
    # создание таблицы с хар-ками
    if category:
        total_count = column.value_counts()
        stats = ['count', 'percent', 'percent_group', 'percent_customer']
        ticks_change = (set_ticks!= None) and (set_ticks in stats)
        
        result = pd.concat(
            [pd.concat(
                [i.value_counts(),
                 (i.value_counts() / total_count.sum()).round(4),
                 i.value_counts(normalize=True).round(4),
                 (i.value_counts() / total_count).round(4)],
                axis=1,
                keys=stats) for i in dfs],
            keys=df_names,
            axis=0
        ).sort_index(level=[0,1])
            
    else:
        result = pd.concat(
            [i.agg(agg_func).to_frame().T.round(4) for i in dfs],
            keys=df_names,
            axis=0
        ).reset_index(level=1, drop=True)
           
    display(result)
    
    fig, axes = plt.subplots(1, 1, squeeze=False, figsize=(12,6))
    plt.subplots_adjust(hspace=0.5, wspace=0.2)
    fig.suptitle(set_title, fontsize=20, y=0.96) 
    ax = sns.histplot(x=column, hue=target, **plot_attr)
    ax.set_ylabel(ylabel)
    if xlabel != None:
        ax.set_xlabel(xlabel)
    #для категориальных признаков, удобно смотреть соотношения на графике
    if ticks_change:
        ticks = [result.loc[(df_names[:2], i), set_ticks].ravel() for i in result.index.get_level_values(1)]
        for j, arr in enumerate(ticks):
            if 'percent' in set_ticks:
                ticks[j] = '/'.join([f'{i:.2%}' for i in arr])
            else:
                ticks[j] = '/'.join([f'{int(i)}' for i in arr])
        ax.set_xticklabels([str(j) + '\n' + ticks[i]
                            for i, j in enumerate(result.index.get_level_values(1))],
                           rotation=ticks_rotation)
    
    plt.show()
    
    if category:
        print('Cтатистический анализ-критерий хи-квадрат, проверка по значению alpha:', alpha)
        print('H0: Распределение ушедших клиентов по выбранному признаку не отличается от оставшихся')
        print('H1: Распределение ушедших клиентов по выбранному признаку отличается от оставшихся')
        chi2, p_final, dof, expected = st.chi2_contingency(pd.concat([dfs[0].value_counts(), dfs[1].value_counts()],
                                                                        axis=1).values)
    else:
        fig, axes = plt.subplots(1, 1, squeeze=False, figsize=(12,6))
        sns.boxplot(data=data_query, x=column, y=target_col)
        plt.show()
        
        print('Односторонний тест Стьюдента, проверка по значению alpha:', alpha)
        print('H0 - среднее двух генеральных совокупностей равны')
        print('H1 - среднее двух генеральных совокупностей не равны')
        p_final = st.ttest_ind(dfs[0], dfs[1], equal_var = False).pvalue / 2
        
    print('P-value: ', p_final)
    if p_final < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу') 
```

Функция `data_split` делит датасет на тренеровочную и тестовую выборку, удаляет ненужные признаки. Дополнительно применяет апсемплинг, создавая новые наблюдения для класса меньшего размера.

```python
global features, target, features_train, features_test, target_test, target_train, cat_col, num_col
```

```python
def data_split(add_cols_drop, upsample=True, k_n=15):
    global features, target, features_train, features_test, target_test, target_train
    global drop_cols, cat_col_raw, num_col_raw, cat_col, num_col
    
    drop_total = list(set(drop_cols + add_cols_drop))

    cat_col = list(set(cat_col_raw).difference(set(drop_total)))
    num_col = list(set(num_col_raw).difference(set(drop_total)))

    features = data.drop(columns=drop_total)
    features[cat_col] = features[cat_col].astype('str') #for catboost
    target = data['customer_left'].astype('int')
    features_train, features_test, target_train, target_test = train_test_split(
        features, target, test_size=0.25, random_state=SEED,
        stratify=target
    )
    
    if upsample:
        SMOTE_up = SMOTENC(categorical_features=features.columns.isin(cat_col), k_neighbors=k_n, random_state=SEED)
        features_train, target_train = SMOTE_up.fit_resample(features_train, target_train)
    
    print('Train shape:', features_train.shape)
    print('Test shape:', features_test.shape)
    print('Train target values:', target_train.value_counts())
    skim(features)
```

## Обучение моделей

### Объявление функций

Функция `create_pipeline` создает пайплайн, ведет поиск гиперпараметров для модели и подбирает лучший порог вероятности для выбранной модели после кросс-валидации.

```python
def create_pipeline(classif_raw, num_trans=None, cat_trans=None, search_type='grid',
                    calibrate_proba='sigmoid',
                    cv=5, set_iter=10, grid = None, fit_params=None):
    result = {}
    trans = []
    
    if num_trans != None:
        trans.append(('num_trans', num_trans, num_col))
    if cat_trans != None:
        trans.append(('cat_trans', cat_trans, cat_col))
    if grid == None:
        grid = {}
    
    if type(calibrate_proba) == str: #for non-catboost models
        classifier = CalibratedClassifierCV(base_estimator=classif_raw, method=calibrate_proba, cv=3)
        final_grid = {}
        for i in grid:
            final_grid['base_estimator__' + i] = grid[i]
    else:
        classifier = classif_raw
        final_grid = grid
    
    if search_type == 'grid':
        model = GridSearchCV(classifier, param_grid=final_grid, scoring='roc_auc', cv=cv, n_jobs=-1)
    elif search_type == 'random':
        model = RandomizedSearchCV(classifier, param_grid=final_grid, scoring='roc_auc', n_iter=set_iter, cv=cv, n_jobs=-1)
        
    step = []
    if (len(trans) > 0):
        step.append(('preprocesser', ColumnTransformer(transformers=trans, remainder='passthrough')))
    step.append(('classifier', model))
    
    pipe = Pipeline(steps=step)
    fit_params = {} if fit_params == None else fit_params
    
    pipe.fit(features_train, target_train, **fit_params)
    
    
    
    if type(calibrate_proba) == float: #for catboost models
        print('Setting new threshold, old:', pipe.named_steps.classifier.best_estimator_.get_probability_threshold())
        pipe.named_steps.classifier.best_estimator_.set_probability_threshold(
            select_threshold(
                pipe.named_steps.classifier.best_estimator_,
                curve=get_roc_curve(
                    pipe.named_steps.classifier.best_estimator_,
                    Pool(features_train, target_train, cat_features=cat_col)
                ),
                FNR=calibrate_proba)
        )
        print('New threshold', pipe.named_steps.classifier.best_estimator_.get_probability_threshold())
        
    model = pipe.named_steps.classifier
    print('Params:', model.best_params_)
    
    result['pipe'] = pipe
    result['model'] = model
    
    result['score'] = test_model(result, features_train, target_train, show=True)
        
    
    return result
```

Функция `test_model` вычисляет результат основной метрики, также используется для сохранения результатов тестирования и вывода информации по полученным результатам.

```python
def test_model(model, feat, targ, show=False, return_result=True):
    
    proba_one = model['pipe'].predict_proba(feat)[:, 1]
    predictions = model['pipe'].predict(feat)
    result = {}
    result['recall'] = recall_score(targ, predictions)
    result['precision'] = precision_score(targ, predictions)
    result['f1'] = f1_score(targ, predictions)
    result['accuracy'] = accuracy_score(targ, predictions)
    result['roc_auc'] = roc_auc_score(targ, proba_one)
    result = pd.DataFrame(result, index=['score'])

    if show:
        display(result)
        print('Confusion_matrix:')
        display(
            pd.DataFrame(confusion_matrix(targ, predictions),
                         columns=['N_pred', 'P_pred'],
                         index=['N_answer', 'P_answer'])
        )

        fpr, tpr, thresholds = roc_curve(targ, proba_one)
        ax = sns.lineplot(x=fpr, y=tpr)
        sns.lineplot(x=[0, 1], y=[0, 1], linestyle='--', ax=ax)
        ax.set_title('ROC curve', fontsize=15)
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
    
    if (return_result):
        return result
```

Во время обучения были использованы те признаки, которые были выделены во время разведочного анализа как важная состовляющая портрета, у них прослеживалась корреляция с целевым признаком, что подтверждает корректность сделанных выводов. Значение тестовой метрики модели равно 0.9366, что выше необходимого значения. В разведочном анализе помимо выделения важных признаков были даны небольшие рекомендации или было выдвинуто предположение, почему этот признак влияет на уход клиента. На обучение слишком сильно влияет признак `duration`, возможно в датасете не хватает дополнительной информации, которая смогла бы помочь объяснить уход клиентов.
