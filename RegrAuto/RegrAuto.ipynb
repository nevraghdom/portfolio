### Удаление дубликатов, обработка пропущенных значений

Функция `print_del` сохраняет количество удаленных наблюдений и причину.

```python
def print_del(amount, reason, cond=None):
    if amount > 0:
        deleted.append({'amount': amount, 'reason': reason})
        print(f'Удалено значений: {amount}, {amount/df_orig[0]:.3%} от всех значений, причина: {reason}')

        if (cond != None):
            return df.query(cond)
    
    return df    
```

Функция `print_na` выводит на экран информацию о количестве пропущенных значений или информацию о количестве уникальных значений в каждом столбце с пропусками.

```python
def print_na(unique=False):
    na_values = df.isna().sum()[lambda x: x > 0]
    if unique: 
        display(
            pd.concat(
                {na_values.index[ind] : df.groupby([i])['date_crawled']
                                          .count().sort_values(ascending=False) for ind, i in enumerate(na_values.index)})
                    .to_frame().rename(columns={'date_crawled': 'amount'})
        )
    else: # total amount of na values in each column
        print(f'Пропущено значений: {na_values.sum()}, {na_values.sum() / (df.shape[0] * df.shape[1]):.2%}')
        if na_values.sum():
            display(
                    pd.DataFrame(
                    np.array([na_values, df[na_values.index].nunique()]).T,
                    columns=['amount', 'nunique'],
                    index= na_values.index
                )
            ) 
```

Функция `distr_plot` выводит на экран распределение признака и распределение указанных подвыборок.

```python
def distr_plot(data, cond, column, plot_func=sns.histplot, **kwargs):
    if type(cond) != list:
        cond = [cond]
    if type(column) != list:
        column = [column] * len(cond)
    
    if len(column) != len(cond):
        return -1
    
    size = 2
    row = ((len(cond) - 1) // size) + 1
    col = (len(cond) > 1) * size
    fig, ax = plt.subplots(row, max(1, col), figsize=(8 * max(1, col), 6 * row), squeeze=False)
    
    for i in range(len(cond)):
        ax_set = ax[i // size][i % size]
        
        if (type(cond[i]) == str):
            plot_func(x=data.query(cond[i])[column[i]], ax=ax_set, **kwargs)
            ax_set.set_title('Distribution ' + cond[i], fontsize=15)
        else:
            plot_func(x=data[column[i]], ax=ax_set, **kwargs)
            ax_set.set_title('Distribution ' + column[i], fontsize=15)
    
    fig.suptitle('Distributions' ,fontsize=25)
    plt.tight_layout()
    plt.show()
```

Функция `data_split` делит датасет на тренировочную и тестовую выборку, удаляет ненужные признаки.

```python
def data_split(cols_drop, upsample=True, k_n=15):
    global features, target, features_train, features_test, target_test, target_train, cat_col, num_col
    
    target = df['price']
    features = df.drop(columns=['price'] + list(set(cols_drop)))
    num_col = list(features.select_dtypes(exclude=['object', 'string']).columns)
    cat_col = list(features.select_dtypes(include=['object', 'string']).columns)
    features[cat_col] = features[cat_col].astype('str')

    features_train, features_test, target_train, target_test =  train_test_split(features, target,
                                                                                 test_size=0.25,
                                                                                 random_state=SEED)
    
    print('Train shape:', features_train.shape)
    print('Test shape:', features_test.shape)
    skim(features)
```

Функция `create_pipeline` создает пайплайн, ведет поиск гиперпараметров для модели и выводит результаты обучения на экран.

```python
def create_pipeline(regressor, num_trans=None, cat_trans=None, search_type='grid',
                    cv=5, set_iter=10, grid = None, fit_params=None):
    start_time = time.time()
    trans = []
    
    if num_trans != None:
        trans.append(('num_trans', num_trans, num_col))
    if cat_trans != None:
        trans.append(('cat_trans', cat_trans, cat_col))
    if grid == None:
        grid = {}
    
    if search_type == 'grid':
        model = GridSearchCV(regressor, param_grid=grid, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    elif search_type == 'random':
        model = RandomizedSearchCV(regressor, param_grid=grid, scoring='neg_root_mean_squared_error', n_iter=set_iter, cv=cv, n_jobs=-1)
        
    step = []
    if (len(trans) > 0):
        step.append(('preprocesser', ColumnTransformer(transformers=trans, remainder='passthrough')))
    step.append(('regressor', model))
    
    pipe = Pipeline(steps=step)
    fit_params = {} if fit_params == None else fit_params
    
    pipe.fit(features_train, target_train, **fit_params)
    
    time_total = round(time.time() - start_time)
    pipe.named_steps.regressor.time_ = time_total
    score = round(abs(model.best_score_))
    pipe.named_steps.regressor.score_ = score
    
    print('Params:', pipe.named_steps.regressor.best_params_)
    print('RMSE:', score)
    print('Time spent:', time_total, 'seconds')
    
    return pipe
```

## Тестирование моделей

Так как еще важна скорость предсказания, проверим все модели. Как и ожидалось, `lgb` показала лучший результат, однако у неё достаточно большое время предсказания. У модели `cat` значение метрики почти такое же, однако скорость предсказания почти в 3 раза быстрее. Подходят обе модели, здесь стоит уточнить у заказчика что предпочтительнее: скорость предсказания или обучения.

```python
models_time = []
idx = []

for key, pipe in models.items():
    model = pipe.named_steps.regressor
    cv_keys = pd.Series(model.cv_results_.keys())
    cv_keys = cv_keys[cv_keys.str.contains(r'split\d')].values
    dict_score = {}
    
    for ind, j in enumerate(cv_keys):
        dict_score['split_' + str(ind)] = round(abs(model.cv_results_[j][0]))
    
    start_time = time.time()
    score = round(mean_squared_error(target_test, pipe.predict(features_test), squared=False), 2)
    
    dict_score['time_pred'] = round(time.time() - start_time, 3)
    dict_score['time_train'] = model.time_
    dict_score['valid_score'] = round(abs(model.score_))
    dict_score['test_score'] = score
    
    models_time.append(dict_score)
    idx.append(key)
    
result = pd.DataFrame(models_time, index = idx).sort_values(by=['test_score'])
result
```
